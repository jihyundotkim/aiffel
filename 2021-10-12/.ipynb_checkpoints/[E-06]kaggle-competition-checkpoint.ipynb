```python
import warnings
warnings.filterwarnings("ignore")

import os
from os.path import join

import pandas as pd
import numpy as np

from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
import xgboost as xgb
import lightgbm as lgb

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
import seaborn as sns

random_state = 1994
```

```python
data_dir = os.getenv('HOME')+'/AIFFEL/2021-10-12'

train_data_path = join(data_dir, 'train.csv')
test_data_path = join(data_dir, 'test.csv')
submission_data_path = join(data_dir, 'sample_submission.csv')
```

필요한 라이브러리와 데이터를 가져옵니다.

```python
train = pd.read_csv(train_data_path)
test = pd.read_csv(test_data_path)
train_id = train['id']
del train['id']
test_id = test['id']
del test ['id']
y = train['price']
del train['price']
del train['date']
del test['date']
```

우선 타겟 데이터인 price를 y에 넣고 train에서는 삭제합니다. date는 2014년, 2015년 둘 중 하나인 것 같고 결과에 크게 영향을 미치지 않을 것 같아 없앴습니다. id 또한 결과에 영향을 주지 않기 때문에 없앴습니다.

```python
for i, row in train.iterrows():
    if row['yr_renovated'] == 0:
        train.at[i,'yr_renovated'] = row['yr_built']
for i, row in test.iterrows():
    if row['yr_renovated'] == 0:
        test.at[i,'yr_renovated'] = row['yr_built']
```

집을 재건축 했을 경우는 재건축 연도인데 재건축을 안 했으면 재건축 연도가 0인 것은 데이터 분포가 맞지 않는 것 같아 재건축하지 않았을 경우 재건축 연도가 처음 건축 연도와 같은 것으로 했습니다.

```python
fig, ax = plt.subplots(9, 2, figsize=(20, 60))
count = 0
columns = train.columns
for row in range(9):
    for col in range(2):
        sns.kdeplot(train[columns[count]], ax=ax[row][col])
        count+=1
        if count == 18 :
            break
```

변수들의 분포를 확인해봅니다.

```python
rs1 = RobustScaler()
rs2 = RobustScaler()
train = rs1.fit_transform(train)
test = rs2.fit_transform(test)
```

scikit-learn의 모듈을 이용해 정규화를 합니다.

```python
fig, ax = plt.subplots(9, 2, figsize=(20, 60))
count = 0
for row in range(9):
    for col in range(2):
        sns.kdeplot(train[count], ax=ax[row][col])
        count+=1
        if count == 18 :
            break
```

좀 더 정규화된 분포를 볼 수 있습니다.

```python
gboost = GradientBoostingRegressor(random_state=random_state, learning_rate=0.0001, max_depth=1, n_estimators=50)
xgboost = xgb.XGBRegressor(random_state=random_state, learning_rate=0.0001, max_depth=1, n_estimators=50)
lightgbm = lgb.LGBMRegressor(random_state=random_state, learning_rate=0.0001, max_depth=1, n_estimators=50)
rdforest = RandomForestRegressor(random_state=random_state, max_depth=1, n_estimators=70)

models = [gboost, xgboost, lightgbm, rdforest]
```

그리드 서치로 찾아낸 최적의 하이퍼파라미터로 인스턴스들을 생성합니다. 그리드 서치는 한 번만 하면 되고 시간이 오래 걸려 코드에서는 뺐습니다.

```python
for m in models: 
    m.fit(train, y)

print(m)

predictions = np.column_stack([m.predict(test) for m in models])

predictions_df = pd.DataFrame(predictions, columns = ['gboost','xgboost','lightgbm', 'rdforest'])
predictions_df.head()
```

```python
y_pred = np.mean(predictions, axis=1)
```

모든 모델의 prediction을 mean으로 블렌딩 해줍니다

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 15))
count = 0
for row in range(2):
    for column in range(2):
        sns.kdeplot(predictions_df[predictions_df.columns[count]], ax=ax[row][column])
        count += 1
```

모든 예측값을 plot 해봅니다.

```python
submission = pd.DataFrame(data={id=test_id, })
submission.to_csv('submission.csv', index=False)
```
